```python
# On renomme les colonnes de chaque d pour plus de clareté : 

liste_dfs = [df_chomage, df_balance_co, df_conf_conso, df_conf_eco, df_ventes_detail, df_pib, df_pmi, df_prod_indu]
liste_noms = ['chomage', 'balance_co', 'conf_conso', 'conf_eco', 'ventes_detail', 'pib', 'pmi', 'prod_indu']

for df, name in zip(liste_dfs, liste_noms):

    # Réinitialiser l’index
    df.reset_index(drop=True, inplace=True)

    # Renommer les colonnes comme demandé
    df.columns = [f'date_{name}', f'note_{name}', f'niveau_{name}']

    # Convertir la colonne date en datetime
    df[f'date_{name}'] = pd.to_datetime(df[f'date_{name}'], errors='coerce')

    # Formater en YYYY-MM
    df[f'date_{name}'] = df[f'date_{name}'].dt.strftime('%Y-%m')
```

```python

# Concaténation des df "propre" pour faire un gros df avec toutes les données
for df, name in zip(liste_dfs, liste_noms):
    df.rename(columns={f'date_{name}': 'DATE'}, inplace=True)

# 2. Initialiser le DataFrame final avec le premier DataFrame de la liste
df_final = liste_dfs[0]

# 3. Faire un merge successif sur la colonne 'DATE'
for df in liste_dfs[1:]:
    df_final = df_final.merge(df, on='DATE', how='outer')

# 4. Trier par date
df_final = df_final.sort_values(by='DATE').reset_index(drop=True)

# 5. Optionnel : vérifier les résultats
df_final=df_final.dropna()
df_final
```

```python
# Ajout du niveau du SP500, pour chacun des mois qui se trouve dans le df (on prend le niveau moyen car on a des données mensuelles)

sp500 = yf.download('^GSPC', start='1999-01-01', end='2025-08-01')

# Calculer la moyenne mensuelle
monthly_avg = sp500['Close'].resample('M').mean()

# Transformer en DataFrame propre
df_mensuel = monthly_avg.reset_index()
df_mensuel.columns = ['DATE', 'SP_avg']

# Garder seulement janvier 1999 à juillet 2025
df_mensuel = df_mensuel[(df_mensuel['DATE'] >= '1999-01-01') & (df_mensuel['DATE'] <= '2025-07-31')]
df_mensuel['DATE'] = pd.to_datetime(df_mensuel['DATE'])
df_mensuel['DATE'] = df_mensuel['DATE'].dt.strftime('%Y-%m')

df_mensuel
```

```python
# df final
df = df_final.merge(df_mensuel, on='DATE', how='left')
df['SP_return']=df['SP_avg'].pct_change()
df=df.dropna()
df
```

2 - Analyse des données

```python
df = df.sort_values("DATE").reset_index(drop=True)

features = [
    "niveau_chomage", "niveau_balance_co", "niveau_conf_conso",
    "niveau_conf_eco", "niveau_ventes_detail", "niveau_pib", "niveau_prod_indu"
]

df["niveau_pmi_Mois_prec"] = df["niveau_pmi"].shift(1) # On crée une nouvelle colonne avec les données du niveau PMI décalé vers le bas

for feature in features:
    df[f"{feature}_Mois_prec"] = df[feature].shift(1)

results = []

for feature in features:

    # Pour chaque feature, on crée le dictionnaire suviant  :
    conds = {
        "PMI < 50": df["niveau_pmi_Mois_prec"] < 50,
        "PMI > 50": df["niveau_pmi_Mois_prec"] >= 50,
        "PMI up" : df["niveau_pmi"] > df["niveau_pmi_Mois_prec"],
        "PMI down" : df["niveau_pmi"] < df["niveau_pmi_Mois_prec"],
        f"{feature} up" : df[feature] > df[f"{feature}_Mois_prec"],
        f"{feature} down" : df[feature] < df[f"{feature}_Mois_prec"]
    }

    for pmi_level in ["PMI < 50", "PMI > 50"]:
        for pmi_change in ["PMI up", "PMI down"]:
            for feat_change in [f"{feature} up", f"{feature} down"]: # Les trois boucles permettent de tester les  8 configurations possibles
                mask = conds[pmi_level] & conds[pmi_change] & conds[feat_change]
                mean_return = df.loc[mask, "SP_return"].mean()

                results.append({
                    "feature": feature,
                    "pmi_level": pmi_level,
                    "pmi_change": pmi_change,
                    "feature_change": feat_change,
                    "mean_return": mean_return,

                })

# Convertir en DataFrame
results_df = pd.DataFrame(results)

```

```python
# Stratégie 1 :  stratégie dans laquelle on n'investit que dans les situations où on a obtenu un redement moyen positif (cf partie résultat)

def compute_signal_simple1(row):

    conditions_defavorables = [
        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_ventes_detail"] > row["niveau_ventes_detail_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_conf_eco"] > row["niveau_conf_eco_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_chomage"] > row["niveau_chomage_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_ventes_detail"] < row["niveau_ventes_detail_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_pib"] > row["niveau_pib_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_chomage"] < row["niveau_chomage_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_balance_co"] < row["niveau_balance_co_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_conf_eco"] < row["niveau_conf_eco_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_prod_indu"] > row["niveau_prod_indu_Mois_prec"]),

        (row["niveau_pmi_Mois_prec"] > 50 and
         row["niveau_pmi"] < row["niveau_pmi_Mois_prec"] and
         row["niveau_conf_conso"] > row["niveau_conf_conso_Mois_prec"])
    ]

    if any(conditions_defavorables):
        return 0
    else:
        return 1

# backtesting stratégie 1
df["invested"] = df.apply(compute_signal_simple1, axis=1)
df["strategy_return"] = df["SP_return"] * df["invested"]
df["cumulative_strategy"] = (1 + df["strategy_return"]).cumprod()
df["cumulative_market"] = (1 + df["SP_return"]).cumprod()

print(f"Nombre de mois investis : {df['invested'].sum()}")
print(f"Nombre de mois totaux : {len(df)}")
print(f"Rendement final stratégie : {df['cumulative_strategy'].iloc[-1]:.2f}x")

import pandas as pd

# Conversion en datetime pour l'axe des abscisses
df["DATE"] = pd.to_datetime(df["DATE"])

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(12,6))
plt.plot(df["DATE"], df["cumulative_market"], label="Marché (S&P 500)", linestyle="--", alpha=0.6)
plt.plot(df["DATE"], df["cumulative_strategy"], label="Stratégie 1: long and hold")
plt.xlabel("Date")
plt.ylabel("Rendement cumulé")
plt.title("Backtest stratégie : N’investir que si pas de conditions défavorables")
plt.legend()

plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.xticks(rotation=45)
plt.show()
```
